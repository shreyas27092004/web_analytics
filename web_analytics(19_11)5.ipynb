```python
import pandas as pd

# Sample clickstream data
data = {
 'user_id': ['U1','U1','U1','U2','U2','U3','U3'],
 'timestamp': [
 '2025-01-01 10:00:00',
 '2025-01-01 10:05:00',
 '2025-01-01 10:40:00',
 '2025-01-01 11:00:00',
 '2025-01-01 11:50:00',
 '2025-01-01 12:00:00',
 '2025-01-01 12:20:00'
 ]
}

df = pd.DataFrame(data)

# Convert timestamp column to datetime
df['timestamp'] = pd.to_datetime(df['timestamp'])

# Sort by user and timestamp
df = df.sort_values(by=['user_id','timestamp'])

# Identify session based on 30-minute inactivity rule
session_threshold = pd.Timedelta(minutes=30)

df['session_id'] = (df.groupby('user_id')['timestamp']
 .diff()
 .gt(session_threshold)
 .cumsum())

# Display results
df
```
